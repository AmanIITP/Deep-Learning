```python
import numpy as np
import pandas as pd
import time
```

```python
df = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
df.head()
```

```python
df=df[['Gender','Age','EstimatedSalary','Purchased']]
df['Gender'].replace(['Male','Female'],[0,1],inplace=True)
```

```python
df.head()
```

```python
X=df.iloc[:,0:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(10,activation='relu',input_dim=3))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',metrics=['accuracy'])
history = model.fit(X_scaled,y,epochs=100,batch_size=250,validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
```

```python
model=Sequential()
model.add(Dense(10,activation='relu',input_dim=3))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',metrics=['accuracy'])
history = model.fit(X_scaled,y,epochs=100,batch_size=1,validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
```
